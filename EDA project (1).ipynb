### Exploratory Data Analysis

### Gender

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(data = df,x = 'Gender',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

```python
ax = sns.countplot(data=df, x='Age Group', hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Number of Orders by Age Group and Gender", fontsize=14, weight='bold')
plt.xlabel("Age Group")
plt.ylabel("Order Count")
plt.show()
```

```python
# Total Amount vs Age Group
sales_age = (
    df.groupby(['Age Group'], as_index=False)['Amount']
    .sum()
    .sort_values(by='Amount', ascending=False)
)

# Pick distinct colors for each Age Group
colors = sns.color_palette("tab10", len(sales_age))

plt.figure(figsize=(10,6))
ax = sns.barplot(x='Age Group', y='Amount', data=sales_age, hue='Age Group')

# Add exact total Amount labels above bars
for container in ax.containers:
    ax.bar_label(container, labels=[f"{v:,.0f}" for v in container.datavalues], fontsize=12)

plt.title("Total Amount vs Age Group", fontsize=16)
plt.xlabel("Age Group", fontsize=14)
plt.ylabel("Total Amount", fontsize=14)
plt.show()
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

### State

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders',hue='State')


plt.title("Number of Orders from Top 10 States", fontsize=14, weight='bold')
plt.xlabel("State")
plt.ylabel("Orders")
plt.show()
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(15,5))
ax = sns.barplot(data=sales_state, x='State', y='Amount', hue='State')

# ✅ Add labels on top of bars
for container in ax.containers:
    ax.bar_label(container, labels=[f"{v:,.0f}" for v in container.datavalues], fontsize=11)

plt.title("Total Sales by Top 10 States", fontsize=16)
plt.xlabel("State", fontsize=14)
plt.ylabel("Total Amount", fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

### Marital Status

```python
# Map 0 → Unmarried, 1 → Married
df['Marital_Status_Label'] = df['Marital_Status'].map({0: "Unmarried", 1: "Married"})

# Set figure size
sns.set(rc={'figure.figsize':(7,5)})

# Countplot with new labels
ax = sns.countplot(data=df, x='Marital_Status_Label',hue='Marital_Status_Label')

# Add count labels
for bars in ax.containers:
    ax.bar_label(bars, fmt='%d')

plt.title("Marital Status Distribution", fontsize=14)
plt.xlabel("Marital Status", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation',hue='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.title(" Occupation by Count", fontsize=16)
plt.xlabel("Occupation", fontsize=14)
plt.ylabel("Occupation", fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector


### Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category',hue='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Product Category by Count", fontsize=14, weight='bold')
plt.xlabel("Product_Category")
plt.ylabel(" Count")
plt.show()
```

```python
# Top 10 product categories by total amount
sales_state = (
    df.groupby(['Product_Category'], as_index=False)['Amount']
    .sum()
    .sort_values(by='Amount', ascending=False)
    .head(10)
)

# Generate a gradient palette (dark → light, highest = darkest)
colors = sns.color_palette("Greens", len(sales_state))

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(
    data=sales_state,
    x='Product_Category',
    y='Amount',
    palette=list(reversed(colors))  # ✅ correctly closed
)

# Add value labels on bars
for container in ax.containers:
    ax.bar_label(container, labels=[f"{v:,.0f}" for v in container.datavalues], fontsize=11)

plt.title("Total Sales by Top 10 Product Categories", fontsize=16)
plt.xlabel("Product Category", fontsize=14)
plt.ylabel("Total Amount", fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
 # Top 10 most sold products
fig1, ax1 = plt.subplots(figsize=(12,7))

# Get top 10 products
top_products = (
    df.groupby('Product_ID')['Orders']
    .sum()
    .nlargest(10)
    .sort_values(ascending=False)
)

# Create a colormap (dark for high → light for low)
colors = plt.cm.Blues(np.linspace(0.3, 1, len(top_products)))
colors = colors[::-1]   # reverse so highest orders = darkest

# Plot with gradient colors
top_products.plot(kind='bar', color=colors, ax=ax1)

# Add labels above bars
for container in ax1.containers:
    ax1.bar_label(
        container,
        labels=[f"{v:,.0f}" for v in container.datavalues],
        fontsize=11
    )

# Titles and labels
ax1.set_title("Top 10 Most Sold Products", fontsize=16)
ax1.set_ylabel("Total Orders", fontsize=14)
ax1.set_xlabel("Product ID", fontsize=14)
plt.xticks(rotation=45)

plt.show()

```


### Conclusion:

Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
